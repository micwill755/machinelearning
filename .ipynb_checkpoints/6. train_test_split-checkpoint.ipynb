```python
import pandas as pd
```

```python
df = pd.read_csv("../data/homeprices_dummy.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis='columns')
```

```python
final
```

```python
X = final.drop(['price'], axis='columns')
X
```

```python
y = final.price
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.predict(X)
```

```python
model.predict([[2800, 0, 1]])
```

```python
model.predict([[3400, 0, 0]])
```

```python
model.score(X, y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = pd.read_csv("../data/homeprices_dummy.csv")
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town', 'area']].values
X
```

```python
y = dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:, 1:]
X
```

```python
model.fit(X, y)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in robbinsville
```

```python
df = pd.read_csv("../data/carprices.csv")
df
```

```python
dfc = df
dfc['Car Model'] = le.fit_transform(dfc['Car Model']) # this will replace car model with numerical data
dfc
```

```python
X_c = dfc[['Car Model', 'Mileage', 'Age(yrs)']].values
X_c
```

```python
y_c = dfc['Sell Price($)'].values
y_c
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct_c = ColumnTransformer( transformers=[('OneHot',OneHotEncoder(), [0])],remainder="passthrough")
X_c = ct_c.fit_transform(X_c)
X_c
```

```python
X_c=X_c[:,1:]
X_c
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_c,y_c)
```

```python
model.predict([[0, 1, 45000, 4]])
```

```python
model.predict([[1, 0, 86000, 7]])
```
